```python
import pandas as pd
```

```python
df = pd.read_excel('1.xlsx')
```

```python
df.head(10)
```

```python
df.replace('no data', 'Nil', inplace=True)
```

```python
df.rename(columns={df.columns[0]: 'Country'}, inplace=True)
```

```python
df_melted = df.melt(id_vars='Country', var_name='Year', value_name='Unemployment Rate (%)')
```

```python
df_melted.head(15)
```

```python
from google.colab import files
output_file = 'unemployment_transformed.csv'
df_melted.to_csv(output_file, index=False)

files.download(output_file)
```

```python
df_1 = pd.read_excel('/content/IMF.xls')
df_1.head()
```

```python
print("Current Columns:")
for idx, col in enumerate(df_1.columns):
    print(f"{idx}: {col}")

start = int(input("Enter the start index of columns to drop: "))
end = int(input("Enter the end index of columns to drop: "))

df_1 = df_1.drop(df_1.columns[start:end+1], axis=1)
```

```python
id_vars = ['Government expenditure, percent of GDP (% of GDP)']

df_1_melted = df_1.melt(id_vars=id_vars, var_name='Year', value_name='GDP Growth')


df_1_melted = df_1_melted[df_1_melted['GDP Growth'] != 'no data']
```

```python
df_1_melted.to_csv('gdp_1921-2023.csv', index=False)
```
